```python
m_train = int(input("traindata size: "))
m_test = int(input("testdata size: "))
n = 1
w0 = float(input("offset: "))
w1 = float(input("slope: "))
```

```python
import random

xtrain = [random.random() * 15 for i in range(m_train)]
yhat_train = [w0 + w1 * xtrain[i] for i in range(m_train)]
ytrain = [yhat_train[i] + random.normalvariate(0, 1) for i in range(m_train)]

xtest = [random.random() * 15 for i in range(m_test)]
yhat_test = [w0 + w1 * xtest[i] for i in range(m_test)]
ytest = [yhat_test[i] + random.normalvariate(0, 1) for i in range(m_test)]
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(xtrain, ytrain, 'o', ms=2, c='blue')
plt.plot(xtrain, yhat_train, c='red')
plt.plot(xtest, ytest, 'o', ms=4, c='green')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#save data in files
path = input('traindata file path: ')
with open(path, 'w') as f:
    print(m_train, file=f)
    print(n, file=f)
    for i in range(m_train):
        print(xtrain[i], ytrain[i], sep='\t', file=f)
    
path = input('testdata file path: ')
with open(path, 'w') as f:
    print(m_test, file=f)
    print(n, file=f)
    for i in range(m_test):
        print(xtest[i], ytest[i], sep='\t', file=f)
```
